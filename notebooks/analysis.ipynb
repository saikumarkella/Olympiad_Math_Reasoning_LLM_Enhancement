```python
# analyzing the json files
import json
def load_json(file_path):
    with open(file_path, 'r') as f:
        data = json.load(f)
    return data

path_1 = "/home/sai/Documents/PERSONAL-PROJECTS/LLMs/Projects/Olympiad_Math_Reasoning_LLM_Enhancement/data/raw/aime.json"
data_1 = load_json(path_1)
print(f"Total entries in aime.json: {len(data_1)}")
```
